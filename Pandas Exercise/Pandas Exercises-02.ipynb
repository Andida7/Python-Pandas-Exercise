# Pandas Exercise: Deleting (Wine)

### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the wine dataset from this [address](https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data). 

```python
url="https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data"
data=pd.read_csv(url,sep=',',names=list(range(1,15)))
data.head()
```

### Step 3. Assign it to a variable called wine

```python
wine=data
wine.head(5)
```

### Step 4. Delete the first, fourth, seventh, nineth, eleventh, thirteenth and fourteenth columns

```python
wine.drop([1,4,7,9,11,13,14],inplace=True,axis=1)
wine.head(5)
```

### Step 5. Assign the columns as below:

The attributes are:  
1) alcohol  
2) malic_acid  
3) alcalinity_of_ash  
4) magnesium  
5) flavanoids  
6) proanthocyanins  
7) hue 

```python
wine=wine.rename(columns={2:'alcohol',3:'malic_acid',5:'alcalinity_of_ash',6:'magnesium',8:'flavanoids',10:'flavanoids',12:'hue'})
wine.head(5)
```

### Step 6. Set the values of the first 3 rows from alcohol as NaN

```python
wine.iloc[0:3,0]=np.nan
wine.head(5)
```

### Step 7. Now set the value of the rows 3 and 4 of magnesium as NaN

```python
wine.iloc[2:4,3]=np.nan
wine.head(4)
```

### Step 8. Fill the value of NaN with the number 10 in alcohol and 100 in magnesium

```python
wine.alcohol= wine.alcohol.fillna(10)

wine.magnesium.fillna(100, inplace = True)
wine.head()
```
