```python
#6. In customer table, create a new variable called “customer value segment” using customer value as follows. -  
#- High Value Segment - > 25000  
#- Medium Value Segment – Between 10000 and 25000  
#- Low Value Segment – less than or equal to 10000 
```

```python
Customers['Cust_value_seg']=pd.cut(Customers['Customer_Value'], bins=[0, 10000,25000,5098715], labels=["Low", "Medium", "high"])
Customers.head()
```

```python
#8. How many days between last purchase date and today?
```

```python
today=pd.datetime.now()
```

```python
argdate=Customers.Recent_Date.max()
year = int(argdate / 10000)
month = int((argdate % 10000) / 100)
day = int(argdate % 100)
last_purchase=pd.datetime(year,month,day)
```

```python
(today-last_purchase).days
```

```python
#9. Calculate percentage of sales by each last city, state and region.
```

```python
per_sales=Customers[['Last_region','Last_state','Last_city','Customer_Value']].groupby(['Last_region','Last_state','Last_city']).sum()
```

```python
per_sales.groupby(level=0).apply(lambda x: x/x.sum() * 100).head(10)
```

```python
#10. What is the count of customers, average number of purchases and average purchase transaction value by last state 
#and city. 
```

```python
Customers[['Last_state','Last_city','Buy_Times','Customer_Value','Customer_ID',]].groupby(['Last_state','Last_city']).agg({'Customer_ID':'count','Customer_Value':'mean','Buy_Times':'mean'}).head()
```
